```python
import re
import parse
import pdfplumber
import pandas as pd
from collections import namedtuple
import warnings
warnings.simplefilter("ignore")
```

```python
#Reading pdf file and connverting it to text format for better extraction
```

```python
ap = "data.pdf"
```

```python
with pdfplumber.open(ap) as pdf:
    page = pdf.pages[1] # type page nummber to be converted
    text = page.extract_text() 
```

```python
print(text)
```

```python
#Extracting name using regular expression and splitting seat number and name along with marks 
```

```python
new_vend_re = re.compile(r'^\d{7} (/[A-Z]|[A-Z]).*')
```

```python
for line in text.split('\n'):
    if new_vend_re.match(line):
        print(line)
```

```python
for line in text.split('\n'):
    if new_vend_re.match(line):
        vend_num, *vend_name = line.split()
        vend_name = ' '.join(vend_name)
print(vend_num)
print(vend_name)
```

```python
data_pattern1 = re.compile(r'(\d{2}) (\d{2}) (\d{2}) (\d{2}) (\d{2}) (\d{2}) (\d{2}) (\b\d{2,3}\b)') 
```

```python
for line in text.split('\n'):
    line = data_pattern1.search(line)
    if line:
        print(line.group(8))
```

```python
Inv = namedtuple('Inv', 'std_num, std_name, DISTRIBUTED_COMPUTING, APPLIED_DATA_SCIENCE, SOCIAL_MEDIA_ANALYTICS, ENVIRONMENTAL_MANAGEMENT, DISTRIBUTED_COMPUTING_LAB, APPLIED_DATA_SCIENCE_LAB, SOCIAL_MEDIA_ANALYTICS_LAB, MAJOR_PROJECT_II ')
```

```python
#Alloting extracted values their respective places  
```

```python
line_items = []
for line in text.split('\n'):
    if new_vend_re.match(line):
        std_num, *std_name = line.split()
        std_name = ' '.join(std_name)    
    
    line = data_pattern1.search(line)
    if line:
        DISTRIBUTED_COMPUTING = line.group(1)
        APPLIED_DATA_SCIENCE = line.group(2)
        SOCIAL_MEDIA_ANALYTICS = line.group(3)
        ENVIRONMENTAL_MANAGEMENT = line.group(4)
        DISTRIBUTED_COMPUTING_LAB = line.group(5)
        APPLIED_DATA_SCIENCE_LAB = line.group(6)
        SOCIAL_MEDIA_ANALYTICS_LAB = line.group(7)
        MAJOR_PROJECT_II = line.group(8)
        
        line_items.append(Inv(std_num, std_name, DISTRIBUTED_COMPUTING, APPLIED_DATA_SCIENCE, SOCIAL_MEDIA_ANALYTICS, ENVIRONMENTAL_MANAGEMENT, DISTRIBUTED_COMPUTING_LAB, APPLIED_DATA_SCIENCE_LAB, SOCIAL_MEDIA_ANALYTICS_LAB, MAJOR_PROJECT_II))
```

```python
df_1 = pd.DataFrame(line_items)
df_1.head()
```

```python
#extracting remaining values such as total, remark and category
```

```python
data_pattern_2 = re.compile(r' (\d{2}) ([PF]*)([RR|RLE|RPV|RCC|ADC].*) (\d{3})')
```

```python
for line in text.split('\n'):
    line = data_pattern_2.search(line)
    if line:
        print(line.group(4))
```

```python
Inv_1 = namedtuple('Inv', 'std_num, std_name, REMAR_K, TOTA_l ')
```

```python
line_items = []
for line in text.split('\n'):
    if new_vend_re.match(line):
        std_num, *std_name = line.split()
        std_name = ' '.join(std_name)    
    
    line = data_pattern_2.search(line)
    if line:
        REMAR_K = line.group(3)
        TOTA_l = line.group(4)
        
        
        line_items.append(Inv_1(std_num, std_name, REMAR_K, TOTA_l ))
```

```python
df_2 = pd.DataFrame(line_items)
df_2.head()
```

```python
#combining both above data frames to get desired dataframe
inner = pd.merge(df_1,df_2)
inner
```

```python
#Extracting the C G GP C*G and total 
```

```python
data_pattern_3 = re.compile(r'(\d+ [A-Z] \d+ \d+) (\d+ [A-Z] \d+ \d+) (\d+ [A-Z] \d+ \d+) (\d+ [A-Z] \d+ \d+) (\d+ [A-Z] \d+ \d+) (\d+ [A-Z] \d+ \d+) (\d+ [A-Z] \d+ \d+) (\d+ [A-Z] \d+ \d+) (\d+) (\d{3} \d+\.\d{2})')
```

```python
for line in text.split('\n'):
    line = data_pattern_3.search(line)
    if line:
        print(line.group(10))
```

```python
Inv_2 = namedtuple('Inv', 'std_num, std_name, DISTRIBUTED_COMPUTING, APPLIED_DATA_SCIENCE, SOCIAL_MEDIA_ANALYTICS, ENVIRONMENTAL_MANAGEMENT, DISTRIBUTED_COMPUTING_LAB, APPLIED_DATA_SCIENCE_LAB, SOCIAL_MEDIA_ANALYTICS_LAB, MAJOR_PROJECT_II, REMAR_K, TOTA_l')
```

```python
line_items = []
for line in text.split('\n'):
    if new_vend_re.match(line):
        std_num, *std_name = line.split()
        std_name = ' '.join(std_name)    
    
    line = data_pattern_3.search(line)
    if line:
        DISTRIBUTED_COMPUTING = line.group(1)
        APPLIED_DATA_SCIENCE = line.group(2)
        SOCIAL_MEDIA_ANALYTICS = line.group(3)
        ENVIRONMENTAL_MANAGEMENT = line.group(4)
        DISTRIBUTED_COMPUTING_LAB = line.group(5)
        APPLIED_DATA_SCIENCE_LAB = line.group(6)
        SOCIAL_MEDIA_ANALYTICS_LAB = line.group(7)
        MAJOR_PROJECT_II = line.group(8)
        REMAR_K = line.group(9)
        TOTA_l = line.group(10)
        
        line_items.append(Inv_2(std_num, std_name, DISTRIBUTED_COMPUTING, APPLIED_DATA_SCIENCE, SOCIAL_MEDIA_ANALYTICS, ENVIRONMENTAL_MANAGEMENT, DISTRIBUTED_COMPUTING_LAB, APPLIED_DATA_SCIENCE_LAB, SOCIAL_MEDIA_ANALYTICS_LAB, MAJOR_PROJECT_II,REMAR_K, TOTA_l))
```

```python
df_3 = pd.DataFrame(line_items)
df_3.head()
```

```python
#Mearging data frame to get required output
```

```python
entry_1= pd.DataFrame()
for i in df_3['std_num']:
    entry = inner.loc[inner['std_num']==i]
    entry=entry.append(df_3[df_3['std_num']==i],ignore_index=True)
    entry_1=entry_1.append(entry)
entry_1
```
